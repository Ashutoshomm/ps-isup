```python
import random as rd
import pandas as pd
import numpy as np
import collections
import seaborn as sns
```

```python
##Numbers from 60 to 70
a=[]
for i in range(60,71):
    a.append(i)
print(a)
```

```python
#1000 random numbers between 60 to 70
Ages=[]
for i  in range(1000):
    nm=(rd.choice(a))
    Ages.append(nm)
    
print(Ages)
```

```python
#Mean of Ages
mean=np.mean(Ages)
print("mean Ages=",mean)
```

```python
#Standar deviation of Ages
stand_dev=np.std(Ages)
print("Standard deviation of Ages=",stand_dev)
```

```python
frequency = collections.Counter(Ages)
print(frequency)
```

```python
import matplotlib.pyplot as plt
plt.hist(Ages)
```

```python
#sample mean of 30 sample Space
sample=[]
Smean=[]
for j in range(500):
    for i in range(30):
        j=rd.choice(Ages)
        sample.append(j)
    s=np.mean(sample)
    Smean.append(s)
       # print(sample)
print(Smean)
```

```python
len(Smean)
```

```python
#Graph
grph1=sns.displot(Smean, kind='kde')
```

```python
# mean of 30 samples
print("mean of Sample mean = ",np.mean(Smean))
```

```python
# Standar Deviation of 30 samples
print('STD = ',np.std(Smean))
```

```python
import math
print(stand_dev/math.sqrt(30))
```
